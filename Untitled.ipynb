Since this is only the page one and there are other pages we will use a loop

```python
#creating an empty data frame
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,400):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=3a9134542a9fe239a3bf2ea122f70d11&language=en-US&page={}'.format(i))
    tempdf = pd.DataFrame(response.json()['results'])[['id','title','adult','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df,tempdf],ignore_index=True)
```

```python
df = df.drop_duplicates()
```

```python
df[df['title'] == 'GoodFellas']
```

```python
df.to_csv('top_rated_movies.csv',index=False)
```
